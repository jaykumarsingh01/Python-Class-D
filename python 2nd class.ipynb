```python
# Comparison operator

a=20
b=30
print(a==b) # Equal
print(a!=b) #  Not Equal
print(a>b)  # Greater
print(a<b)  # Less
print(a>=b) # Greater or Equal
print(a<=b) # Less or Equal
```

```python
a=20
b=30
c=40
print(a==b==c) # Equal
print(a!=b!=c) #  Not Equal
print(a>b>c)  # Greater
print(a<b<c)  # Less
print(a>=b>=c) # Greater or Equal
print(a<=b<=c) # Less or Equal
```

```python
# Logical Operator
x = True
y = False

print(x and y)   # Both True
print(x or y)    # At least one True
print(not x)     # Reverse Boolean

age = 25
salary = 50000

print(age > 18 and salary > 30000)
```

```python
age = 20
marks =85
course = True
result = (age<15 & marks> 90) and Course
print(result)
```

```python
# Assignment OPerator
x=10 
x+=5  # x=x+5
x-=2  # x=x-2
x*=3  # x=x*3
x/=2  # x=x/2
print(x)
```

```python
# membership operation (in,not in)
name= 'python'
print("p" in name)  # true
print("Z" not in name)  # true
```

# String Slicing


```python
m="MachineLearning"
print(m[0:7])
print(m[7:])
print(m[:7])
print(m[:]) # Entire string
print(m[5:8])
print(m[::-1])
```

# Multiple Value

```python
def k (a,b):
    return a+b, a*b

add , mul = k(30,3)
print(add)
print(mul)
    
```

```python
def k (a,b,c,d):
    return a+b+c+d, a*b*c*d

add , mul = k(3,3,3,3)
print(add)
print(mul)
```

# Maximum and Minimum

```python
def a (a,b,c,d,e):
    return max (a,b,c,d,e)

print(a(10,20,30,40,50))
```

```python
def a (a,b,c,d,e):
    return min (a,b,c,d,e)

print(a(10,20,30,40,50))
```
